```python
import pickle

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import auc, roc_curve, roc_auc_score, f1_score, precision_recall_curve
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_text

pd.__version__
```

Now let's look at the correlations between variables. To avoid overwhelming the correlation plot, low correlations will be filtered.

```python
threshold = 0.3

corr = df.corr()
corr = corr.where((np.abs(corr) >= threshold))
np.fill_diagonal(corr.values, np.nan)
corr = corr.dropna(how="all").dropna(axis=1, how="all")
np.fill_diagonal(corr.values, 1)

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, vmin=-1, vmax=1, mask=corr.isnull())

plt.show()
```

As we can see, there are no pairs of highly correlated variables, so the linear models will not be unstable due to this

### Preparing data for training

```python
y = 'Diabetes'

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

X_train = df_train.loc[:, df.columns != y].copy()
X_val = df_val.loc[:, df.columns != y].copy()
X_test = df_test.loc[:, df.columns != y].copy()

y_train = df_train[y].values
y_val = df_val[y].values
y_test = df_test[y].values

for d in [X_train, X_val, X_test]:
    print(round(len(d)/len(df),2))
```

A simple analysis of feature importance. We will only the training set for this purpose

```python
plt.figure(figsize=(5,5))
plt.plot([0,1], [0,1], label='Random', linestyle='--')

num_cols = X_train.columns[X_train.dtypes == np.float64]

auc_values = {}

for col in num_cols:
    fpr, tpr, thresholds = roc_curve(y_train, X_train[col])
    auc_value = auc(fpr, tpr)

    auc_values[col] = auc_value
    
    if auc_value < 0.5:
        fpr, tpr, thresholds = roc_curve(y_train, -X_train[col])
        auc_value = auc(fpr, tpr)
    label = f'{col} ({round(auc_value,2)})'
    plt.plot(fpr, tpr, label=label)

auc_values = {k: v for k, v in sorted(auc_values.items(), key=lambda item: item[1], reverse=True)}
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```

### Training

#### Baseline model (logistic regression)

```python
model_baseline = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=1)
model_baseline.fit(X_train, y_train)

y_pred = model_baseline.predict_proba(X_val)[:, 1]

fpr, tpr, thresholds = roc_curve(y_val, y_pred)
auc_value = auc(fpr, tpr)
f1_value = f1_score(y_val, model_baseline.predict(X_val))

print(f'AUC:\t{round(auc_value, 3)}')
print(f'F1:\t{round(f1_value, 3)}')

plt.figure(figsize=(5,5))
plt.plot([0,1], [0,1], label='Random', linestyle='--')
plt.plot(fpr, tpr)
plt.show()
```

#### Grid search (logistic regression)

```python
model_logreg = LogisticRegression(solver='liblinear', max_iter=1000, random_state=1)

param_grid = {
    # 'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'C': [0.01, 0.1, 0.5, 1, 1.5, 2],
    'penalty': ['l1', 'l2']  # Penalty type
}

grid_search = GridSearchCV(estimator=model_logreg, param_grid=param_grid, cv=5, scoring='f1', verbose=1, n_jobs=4)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)

print("Best Hyperparameters:", grid_search.best_params_)
print("Best Cross-Validation F1:", grid_search.best_score_)

print("Valid Set Accuracy:", accuracy_score(y_val, y_pred))
print("Classification Report:")
print(classification_report(y_val, y_pred))
```

#### Decision tree (baseline)

```python
clf = DecisionTreeClassifier(max_depth=8)
clf = clf.fit(X_train, y_train)
# tree_text = export_text(clf, feature_names=list(dv.get_feature_names_out()))

tree_text = export_text(clf)

y_pred_proba = clf.predict_proba(X_val)[:, 1]

# with open('decision_tree_depth8.pkl', 'wb') as file:
#     pickle.dump(clf, file)

fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
auc_value = auc(fpr, tpr)
f1_value = f1_score(y_val, model_baseline.predict(X_val))

precision, recall, thresholds = precision_recall_curve(y_val, y_pred_proba)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10) # 1e-10 is here to avoid division by zero
best_threshold = thresholds[np.argmax(f1_scores)]

y_pred_binary = [1 if y > best_threshold else 0 for y in y_pred_proba]
f1_thr = f1_score(y_val, y_pred_binary)

print(f'AUC:\t{round(auc_value, 3)}')
print(f'F1:\t{round(f1_value, 3)}')
print(f'F1 thr:\t{round(f1_thr, 3)}')
print(f"Best threshold (F1-Score): {round(best_threshold, 3)}")


plt.figure(figsize=(5,5))
plt.plot([0,1], [0,1], label='Random', linestyle='--')
plt.plot(fpr, tpr)
plt.show()
```

```python
model_dtree = DecisionTreeClassifier(random_state=1)

param_grid_dtree = {
    'max_depth': [6, 7, 8, 9, 10],
    'min_samples_split': [2, 10],
    'min_samples_leaf': [1, 5]
}

grid_search = GridSearchCV(estimator=model_dtree, param_grid=param_grid_dtree, cv=5, scoring='f1', verbose=1, n_jobs=4)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)

print("Best Hyperparameters:", grid_search.best_params_)
print("Best Cross-Validation F1:", grid_search.best_score_)

print("Valid Set Accuracy:", accuracy_score(y_val, y_pred))
print("Classification Report:")
print(classification_report(y_val, y_pred))
```
